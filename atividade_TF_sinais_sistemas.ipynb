# ***Dados***

$$x(t) = 4\text{rect}(200t)$$

$$\frac{dy(t)}{dt} + 1000y(t) = 1000x(t)$$

$$y(t) = ?$$

* **Objetivo:** Determinar $y(t)$.

* **Problema:** $y(t)$ é definida por $y(t) = x(t) * h(t)$, ou seja, é necessário convoluir os 2 sinais. O sinal $x(t)$ é um bom sinal para convolução (pulso retangular), mas $h(t)$ é uma equação diferencial, o que torna o processo mais trabalhoso.

* **Solução:** Levar para o domínio da frequência, a fim de que a derivada em $h(t)$ vire uma multiplicação e a inversa seja uma função mais fácil de se trabalhar do que a equação diferencial.


-------------------------------------------------------------------------


# ***Aplicando a definição da Transformada de Fourier em $x(t)$:***

* Seja a Transformada de Fourier para uma função contínua e integrável definida por:

$$X(\omega) = \int_{-\infty}^{\infty} x(t) e^{-j\omega t} dt$$

* Seja o sinal $x(t)$ definido por:

$$x(t) = 4\text{rect}(200t)$$

* Onde podemos escrever da seguinte forma:

$$x(t) = \begin{cases} 4, & -\frac{1}{400} < t < \frac{1}{400} \\ 0, & \text{fora} \end{cases}$$

* ***Para $t < -\frac{1}{400}$:***

$$\boxed{X(\omega) = \int_{-\infty}^{-1/400} 0 \cdot e^{-j\omega t} dt = 0}$$

* ***Para $-\frac{1}{400} < t < \frac{1}{400}$:***

$$X(\omega) = \int_{-1/400}^{1/400} 4 e^{-j\omega t} dt$$

$$X(\omega) = \left[ \frac{4 e^{-j\omega t}}{-j\omega} \right]_{-1/400}^{1/400}$$

* Aplicando os limites:

$$X(\omega) = \left( \frac{-4}{j\omega} \right) (e^{-j\omega \frac{1}{400}} - e^{-j\omega (\frac{-1}{400})})$$

$$X(\omega) = \left( \frac{-4}{j\omega} \right) \cdot (-1) (e^{j\omega \frac{1}{400}} - e^{-j\omega \frac{1}{400}})$$

$$X(\omega) = \left( \frac{4}{j\omega} \right) (e^{j\omega \frac{1}{400}} - e^{-j\omega \frac{1}{400}})$$

$$X(\omega) = \frac{4}{j\omega} (e^{j\omega \frac{1}{400}} - e^{-j\omega \frac{1}{400}})$$

$$\boxed{X(\omega) = \frac{4}{j\omega} (e^{j\omega \ 0,0025} - e^{-j\omega \ 0,0025})}$$

* ***Para $t > \frac{1}{400}$:***

$$\boxed{X(\omega) = \int_{1/400}^{\infty} 0 \cdot e^{-j\omega t} dt = 0}$$


-------------------------------------------------------------------------

# ***Transformando o impulso $h(t)$:***

$$\frac{dy(t)}{dt} + 1000y(t) = 1000x(t)$$

* Pela propriedade da diferenciação no tempo, a derivada equivale à multiplicação por $j\omega$:

$$j\omega Y(\omega) + 1000Y(\omega) = 1000X(\omega)$$

$$Y(\omega)(j\omega + 1000) = 1000X(\omega)$$

$$\boxed{Y(\omega) = \frac{1000}{j\omega + 1000} \cdot X(\omega)}$$

* Onde $\frac{1000}{j\omega + 1000}$ é o $H(\omega)$.

$$Y(\omega) = H(\omega) \cdot X(\omega)$$

* Substituindo os termos $H(\omega)$ e $X(\omega)$:

$$Y(\omega) = \left( \frac{1000}{j\omega + 1000} \right) \left( \frac{4}{j\omega} (e^{j\omega 0,0025} - e^{-j\omega 0,0025}) \right)$$

* Dessa forma, obtemos a saída $Y(\omega)$ (em frequência).

-------------------------------------------------------------------------

# ***Analisando caminhos***
- Neste ponto, podemos encontrar a inversa de $H(\omega)$, uma vez que, pela tabela, sabemos que ela corresponde a uma exponencial começando a partir do 0 (uma função bem mais simples do que a equação inicial). Embora tenhamos encontrado $Y(\omega)$, a intenção inicial é determinar a saída $y(t)$. A volta pode ser feita de algumas formas, mas 3 caminhos serão abordados neste documento.

-------------------------------------------------------------------------

# ***1º Caminho: Aplicar a transformada inversa nos sinais e convoluir***

* Dado que temos $X(\omega)$ e $H(\omega)$, podemos aplicar a transformada inversa em ambos, voltando para o domínio do tempo. Como dito, a inversa dos 2 sinais são funções conhecidas: $X(\omega)$ é simplesmente a função rect (um pulso retangular) e $H(\omega)$ é uma exponencial (partindo do 0) multiplicada por uma constante.

* Dessa forma:

$$x(t) = \mathcal{F}^{-1}\{X(\omega)\} = 4\text{rect}(200t)$$

$$h(t) = \mathcal{F}^{-1}\{H(\omega)\} = 1000e^{-1000t}u(t)$$

* Ou seja:

$$\boxed{y(t) = 4\text{rect}(200t) * [1000e^{-1000t}u(t)]}$$

* Onde:

$$x(t) = \begin{cases} 4, & -\frac{1}{400} < t < \frac{1}{400} \\ 0, & \text{fora} \end{cases}$$

$$h(t) = \begin{cases} 1000e^{-1000t}, & t \ge 0 \\ 0, & t < 0 \end{cases}$$

* Neste método, usaremos a definição da convolução: $$y(t) = \int_{-\infty}^{\infty} x(\tau)h(t-\tau)d\tau$$

* $x(t)$ será fixa e $h(t)$ variável:

$$x(\tau) = \begin{cases} 4, & -\frac{1}{400} < \tau < \frac{1}{400} \\ 0, & \text{fora} \end{cases}$$

$$h(t - \tau) = \begin{cases} 1000e^{-1000(t - \tau)}, & \tau \le t \\ 0, & \tau > t \end{cases}$$

* ***Para $t < -\frac{1}{400}$ o pulso não começou:***

$$\boxed{y(t) = 0}$$

* ***Para $-\frac{1}{400} < t < \frac{1}{400}$, $h(t)$ está sobrepondo $x(t)$ até certo ponto ($t$):***

$$y(t) = \int_{-1/400}^{t} 4 \cdot 1000 e^{-1000(t-\tau)} d\tau$$

$$y(t) = 4 \cdot 1000 \int_{-1/400}^{t} e^{-1000(t-\tau)} d\tau$$

$$y(t) = 4000 \int_{-1/400}^{t} e^{-1000t} \cdot e^{1000\tau} d\tau$$

* $e^{-1000t}$ não depende de $\tau$:

$$y(t) = 4000 e^{-1000t} \int_{-1/400}^{t} e^{1000\tau} d\tau$$

$$y(t) = 4000 e^{-1000t} \left[ \frac{e^{1000\tau}}{1000} \right]_{-1/400}^{t}$$

- Aplicando os limites:

$$y(t) = 4000 e^{-1000t} \left( \frac{e^{1000t}}{1000} - \frac{e^{1000(-1/400)}}{1000} \right)$$

$$y(t) = 4000 e^{-1000t} \left( \frac{e^{1000t} - e^{-2.5}}{1000} \right)$$

$$y(t) = \frac{4000}{1000} e^{-1000t} (e^{1000t} - e^{-2.5})$$

$$y(t) = 4 e^{-1000t} (e^{1000t} - e^{-2.5})$$

- Fazendo a distributiva:

$$y(t) = 4 \left[ (e^{-1000t} \cdot e^{1000t}) - (e^{-1000t} \cdot e^{-2.5}) \right]$$

$$y(t) = 4 (e^{0} - e^{-(1000t + 2.5)})$$

- Colocando $-1000$ em evidência:

$$\boxed{y(t) = 4 (1 - e^{-1000(t + 0.0025)})}$$

* ***Para $t > \frac{1}{400}$, $h(t)$ já "cobriu" todo $x(t)$:***

$$y(t) = \int_{-1/400}^{1/400} 4 \cdot 1000 e^{-1000(t-\tau)} d\tau$$

$$y(t) = 4000 e^{-1000t} \int_{-1/400}^{1/400} e^{1000\tau} d\tau$$

$$y(t) = 4000 e^{-1000t} \left[ \frac{e^{1000\tau}}{1000} \right]_{-1/400}^{1/400}$$

* Aplicando os limites:

$$y(t) = 4000 e^{-1000t} \left( \frac{e^{1000 \cdot \frac{1}{400}} - e^{1000 \cdot (\frac{-1}{400})}}{1000} \right)$$

$$y(t) = 4000 e^{-1000t} \left( \frac{e^{2.5} - e^{-2.5}}{1000} \right)$$

$$y(t) = 4 e^{-1000t} (e^{2.5} - e^{-2.5})$$

$$y(t) = 4 \left[ (e^{-1000t} \cdot e^{2.5}) - (e^{-1000t} \cdot e^{-2.5}) \right]$$

$$y(t) = 4 (e^{-(1000t - 2.5)} - e^{-(1000t + 2.5)})$$

* Colocando $-1000$ em evidência:

$$\boxed{y(t) = 4 (e^{-1000(t - 0.0025)} - e^{-1000(t + 0.0025)})}$$

* Ao resolver as integrais, chegamos na saída $y(t)$ como uma função definida por partes:

$$y(t) = \begin{cases}
0, & t < -0,0025 \\
4(1 - e^{-1000(t + 0,0025)}), & -0,0025 \le t < 0,0025 \\
4(e^{-1000(t - 0,0025)} - e^{-1000(t + 0,0025)}), & t \ge 0,0025
\end{cases}$$

# ***2º Caminho - Manipular $Y(\omega)$ algebricamente***

- Podemos ir quebrando $Y(\omega)$ em expressões cujas inversas resultem em funções simples. Vamos começar manipulando pela parte mais à esquerda:

$$Y(\omega) = \left( \frac{1000}{j\omega + 1000} \right) \left( \frac{4}{j\omega} (e^{j\omega 0,0025} - e^{-j\omega 0,0025}) \right)$$

* Colocando as constantes em evidência e agrupando os termos de frequência:

$$Y(\omega) = 4 \cdot 1000 \left( \frac{1}{j\omega + 1000} \right) \left( \frac{1}{j\omega} \right) (e^{j\omega 0,0025} - e^{-j\omega 0,0025})$$

$$Y(\omega) = 4000 \left( \frac{1}{j\omega(j\omega + 1000)} \right) (e^{j\omega 0,0025} - e^{-j\omega 0,0025})$$

# ***Resolvendo o primeiro parêntese através de frações parciais:***

$$\frac{1}{j\omega(j\omega + 1000)} = \frac{A}{j\omega} + \frac{B}{j\omega + 1000}$$

* Onde:

$$A = \frac{1}{1000} = 0,001$$

$$B = \frac{1}{-1000} = -0,001$$

* Substituindo A e B:

$$\frac{0,001}{j\omega} + \frac{-0,001}{j\omega + 1000}$$

- Colocando 0,001 em evidência:

$$\boxed{0,001 \left( \frac{1}{j\omega} - \frac{1}{j\omega + 1000} \right)}$$

- Substituindo de volta em $Y(\omega)$:

$$Y(\omega) = 4000 \left[ 0,001 \left( \frac{1}{j\omega} - \frac{1}{j\omega + 1000} \right) \right] (e^{j\omega 0,0025} - e^{-j\omega 0,0025})$$

$$Y(\omega) = 4000 \cdot 0,001 \left( \frac{1}{j\omega} - \frac{1}{j\omega + 1000} \right) (e^{j\omega 0,0025} - e^{-j\omega 0,0025})$$

$$\boxed{Y(\omega) = 4 \left( \frac{1}{j\omega} - \frac{1}{j\omega + 1000} \right) (e^{j\omega 0,0025} - e^{-j\omega 0,0025})}$$

* A partir deste ponto, vamos dividir em duas partes: lado esquerdo = $Y_{base}(\omega)$ e lado direito = $G_{desloc}(\omega)$:

$$Y(\omega) = \underbrace{4 \left( \frac{1}{j\omega} - \frac{1}{j\omega + 1000} \right)}_{Y_{base}(\omega)} \cdot \underbrace{(e^{j\omega 0,0025} - e^{-j\omega 0,0025})}_{G_{desloc}(\omega)}$$

# ***Pausa - Provando que termo impulsivo é nulo em $X(\omega)$:***

- Para que esse caminho dê certo, a inversa $u(t)$ deve ser somente $\frac{1}{j\omega}$, e não $\frac{1}{j\omega} + \pi\delta(\omega)$. Vamos provar que, neste caso, este impulso é 0 em $X(\omega)$. Sabemos que $x(t)$ é:

$$x(t) = 4\text{rect}(200t)$$

* Podemos escrever como:

$$x(t) = \begin{cases} 4, & -\frac{1}{400} < t < \frac{1}{400} \\ 0, & \text{fora} \end{cases}$$

* Mas podemos escrever usando degraus unitários:

$$x(t) = 4\left( u(t + \frac{1}{400}) - u(t - \frac{1}{400}) \right)$$

* Ou na forma decimal:

$$\boxed{x(t) = 4\left( u(t + 0,0025) - u(t - 0,0025) \right)}$$

* Considere 2 propriedades:
    1.  **Transformada do degrau:** $u(t) \leftrightarrow \pi\delta(\omega) + \frac{1}{j\omega}$
    2.  **Deslocamento no tempo:** $x(t - t_0) \leftrightarrow X(\omega)e^{-j\omega t_0}$

* Com base nisso, podemos reescrever $x(t)$ no domínio da frequência:
    * $u(t - \frac{1}{400}) \leftrightarrow (\pi\delta(\omega) + \frac{1}{j\omega})e^{-j\omega \frac{1}{400}}$
    * $u(t + \frac{1}{400}) \leftrightarrow (\pi\delta(\omega) + \frac{1}{j\omega})e^{j\omega \frac{1}{400}}$

* Substituindo em $X(\omega)$:

$$X(\omega) = 4 \left[ \left(\pi\delta(\omega) + \frac{1}{j\omega}\right)e^{j\omega \frac{1}{400}} - \left(\pi\delta(\omega) + \frac{1}{j\omega}\right)e^{-j\omega \frac{1}{400}} \right]$$

* Aplicando a distributiva:

$$X(\omega) = 4 \left[ \pi\delta(\omega)e^{j\omega \frac{1}{400}} + \frac{1}{j\omega}e^{j\omega \frac{1}{400}} - \pi\delta(\omega)e^{-j\omega \frac{1}{400}} - \frac{1}{j\omega}e^{-j\omega \frac{1}{400}} \right]$$

$$X(\omega) = 4 \left[ \left(\pi\delta(\omega)e^{j\omega \frac{1}{400}} - \pi\delta(\omega)e^{-j\omega \frac{1}{400}}\right) + \left(\frac{1}{j\omega}e^{j\omega \frac{1}{400}} - \frac{1}{j\omega}e^{-j\omega \frac{1}{400}}\right) \right]$$

* Colocando $\pi\delta(\omega)$ e $\frac{1}{j\omega}$ em evidência:

$$X(\omega) = 4 \left[ \pi\delta(\omega)(e^{j\omega \frac{1}{400}} - e^{-j\omega \frac{1}{400}}) + \frac{1}{j\omega}(e^{j\omega \frac{1}{400}} - e^{-j\omega \frac{1}{400}}) \right]$$

* Fazendo a distributiva do 4:

$$X(\omega) = 4\pi\delta(\omega)(e^{j\omega \frac{1}{400}} - e^{-j\omega \frac{1}{400}}) + \frac{4}{j\omega}(e^{j\omega \frac{1}{400}} - e^{-j\omega \frac{1}{400}})$$

* Vamos olhar para o lado esquerdo:

$$X(\omega) = \underbrace{4\pi\delta(\omega)(e^{j\omega \frac{1}{400}} - e^{-j\omega \frac{1}{400}})}_{\text{parte impulso}} + \frac{4}{j\omega}(e^{j\omega \frac{1}{400}} - e^{-j\omega \frac{1}{400}})$$

* Pela definição, o $\delta(\omega)$ só é diferente de 0 em $\omega=0$.

$$X_{impulso}(\omega) = 4\pi\delta(\omega)(e^{j\omega \frac{1}{400}} - e^{-j\omega \frac{1}{400}})$$

$$X_{impulso}(0) = 4\pi\delta(0)(e^{0} - e^{0})$$

$$X_{impulso}(0) = 4\pi\delta(0)(1 - 1)$$

$$\boxed{X_{impulso}(0) = 0}$$

* Dessa forma, o termo com impulso é nulo, restando apenas a segunda parte compondo $X(\omega)$, a qual coincide exatamente com a que vimos antes.

# ***Continuando - Aplicando a transformada inversa no termo base:***

* Termo base em frequência:

$$Y_{base}(\omega) = 4 \left( \frac{1}{j\omega} - \frac{1}{j\omega + 1000} \right)$$

* Aplicando a transformada inversa ($\mathcal{F}^{-1}$):
    * Determinamos que $\frac{1}{j\omega} \leftrightarrow u(t)$
    * E que $\frac{1}{j\omega + 1000} \leftrightarrow e^{-1000t}u(t)$

$$y_{base}(t) = \mathcal{F}^{-1} \{ Y_{base}(\omega) \}$$

$$y_{base}(t) = 4( u(t) - e^{-1000t}u(t) )$$

* Colocando $u(t)$ em evidência no final:

$$\boxed{y_{base}(t) = 4( 1 - e^{-1000t} )u(t)}$$

# ***Aplicando transformada inversa no termo com deslocamentos:***

- Retomando o termo com deslocamentos $G_{desloc}(\omega)$:

$$G_{desloc}(\omega) = e^{j\omega 0,0025} - e^{-j\omega 0,0025}$$

* Seja a inversa: $e^{-j\omega t_0} \leftrightarrow \delta(t - t_0)$

$$g_{desloc}(t) = \mathcal{F}^{-1} \{ e^{j\omega 0,0025} - e^{-j\omega 0,0025} \}$$

$$\boxed{g_{desloc}(t) = \delta(t + 0,0025) - \delta(t - 0,0025)}$$

# ***Substituindo e formando em $y(t)$:***

* Sabemos que a multiplicação na frequência equivale à convolução no tempo:

$$\mathcal{F}^{-1}\{Y_{base}(\omega) \cdot G_{desloc}(\omega)\} \iff y_{base}(t) * g_{desloc}(t)$$

* Portanto, a saída no tempo é a convolução do termo base com os impulsos:

$$y(t) = y_{base}(t) * g_{desloc}(t)$$

$$y(t) = [4(1 - e^{-1000t})u(t)] * [\delta(t + 0,0025) - \delta(t - 0,0025)]$$

- Vamos usar a seguinte notação:

$$y(t) = y_{base}(t) * [\delta(t + 0,0025) - \delta(t - 0,0025)]$$

* Aplicando a propriedade distributiva da convolução:

$$y(t) = [y_{base}(t) * \delta(t + 0,0025)] - [y_{base}(t) * \delta(t - 0,0025)]$$

* A convolução de um sinal com um impulso é o próprio sinal deslocado:

$$x(t) * \delta(t - t_0) = x(t - t_0)$$

* Ou seja, o resultado da convolução é o $y_{base}$ deslocado:

$$y(t) = y_{base}(t + 0,0025) - y_{base}(t - 0,0025)$$

- Aplicando em $y_{base}(t)$:

$$y(t) = \left[ 4(1 - e^{-1000(t + 0,0025)})u(t + 0,0025) \right] - \left[ 4(1 - e^{-1000(t - 0,0025)})u(t - 0,0025) \right]$$

- Vamos identificar as partes A e B:

$$y(t) = \underbrace{\left[ 4(1 - e^{-1000(t + 0,0025)})u(t + 0,0025) \right]}_{A} - \underbrace{\left[ 4(1 - e^{-1000(t - 0,0025)})u(t - 0,0025) \right]}_{B}$$

- Visto que a parte A começa em $t > -0,0025$ e a parte B começa em $t > 0,0025$, vamos analisar os intervalos:

- ***Para $t < -0,0025$ as duas partes não começaram (os degraus são 0):***

$$\boxed{y(t) = 0}$$

- ***Para $-0,0025 < t < 0,0025$ somente a parte A não é nula:***

$$y(t) = \left[ 4(1 - e^{-1000(t + 0,0025)}) \right] - 0$$

$$\boxed{y(t) = 4(1 - e^{-1000(t + 0,0025)})}$$

- ***Para $t > 0,0025$ as duas partes não são nulas:***

$$y(t) = \left[ 4(1 - e^{-1000(t + 0,0025)}) \right] - \left[ 4(1 - e^{-1000(t - 0,0025)}) \right]$$

$$y(t) = 4 - 4e^{-1000(t + 0,0025)} - 4 + 4e^{-1000(t - 0,0025)}$$

$$y(t) = 4e^{-1000(t - 0,0025)} - 4e^{-1000(t + 0,0025)}$$

$$\boxed{y(t) = 4(e^{-1000(t - 0,0025)} - e^{-1000(t + 0,0025)})}$$

- Ou seja, podemos reescrever $y(t)$ da mesma forma que encontramos no caminho 1:

$$y(t) = \begin{cases}
0, & t < -0,0025 \\
4(1 - e^{-1000(t + 0,0025)}), & -0,0025 \le t < 0,0025 \\
4(e^{-1000(t - 0,0025)} - e^{-1000(t + 0,0025)}), & t \ge 0,0025
\end{cases}$$

# ***3º Caminho - Usar $x(t)$ como degraus***

- A terceira forma de chegar à saída $y(t)$ é representar $x(t)$ como uma soma de degraus e convoluir com $h(t)$ utilizando a propriedade da integração:

$$x_1(t) * u(t) = \int_{-\infty}^{t} x_1(\tau) d\tau$$

* Retomando $x(t)$:

$$x(t) = 4u(t + 0,0025) - 4u(t - 0,0025)$$

* Retomando a convolução:

$$y(t) = x(t) * h(t)$$

$$y(t) = [4u(t + 0,0025) - 4u(t - 0,0025)] * [1000e^{-1000t}u(t)]$$

- Aplicando a distributiva:

$$y(t) = [4u(t + 0,0025) * 1000e^{-1000t}u(t)] - [4u(t - 0,0025) * 1000e^{-1000t}u(t)]$$

* Multiplicando as constantes:

$$y(t) = [4000 u(t + 0,0025) * e^{-1000t}u(t)] - [4000 u(t - 0,0025) * e^{-1000t}u(t)]$$

$$y(t) = 4000 \left[ (u(t + 0,0025) * e^{-1000t}u(t)) - (u(t - 0,0025) * e^{-1000t}u(t)) \right]$$

- **IMPORTANTE:** Irei omitir $u(t)$ nas exponenciais para deixar $y(t)$ visualmente mais limpa, mas não nos esqueçamos que este degrau é responsável por fazer $h(t)$ iniciar a partir do 0.

$$\boxed{y(t) = 4000 \left[ (u(t + 0,0025) * e^{-1000t}) - (u(t - 0,0025) * e^{-1000t}) \right]}$$

- Podemos fazer a convolução do jeito que está, ou de forma mais direta, simplesmente convoluindo $e^{-1000t}$ com $u(t)$ e aplicando o deslocamento em seguida, através da seguinte propriedade: Seja $y_1(t)$ a convolução de um sinal qualquer com o degrau unitário:

$$y_1(t) = x_1(t) * u(t) = \int_{-\infty}^{\infty} x_1(\tau)u(t-\tau) d\tau$$

* E $y_2(t)$ a convolução com o degrau deslocado:

$$y_2(t) = x_1(t) * u(t-t_0) = \int_{-\infty}^{\infty} x_1(\tau)u((t-t_0)-\tau) d\tau$$

- Podemos chamar $(t-t_0) = \lambda$:

$$y_2(t) = \int_{-\infty}^{\infty} x_1(\tau)u(\lambda-\tau) d\tau$$

- Note que $y_2$ tem a mesma estrutura de $y_1$, mas aplicada em $\lambda$. Logo:

$$y_2(t) = y_1(\lambda)$$

$$y_2(t) = y_1(t - t_0)$$

- Ou seja, calcular a convolução de $x_{qualquer}(t)$ com $u(t-t_0)$ é a mesma coisa que calcular a convolução de $x_{qualquer}(t)$ com $u(t)$ e aplicar o deslocamento $t_0$ posteriormente. Isso facilita nosso trabalho, sendo necessário fazer somente uma integral. Aplicando ao nosso caso, vamos definir um $y_{coringa}(t)$:

$$y_{coringa}(t) = e^{-1000t} * u(t)$$

$$y_{coringa}(t) = \int_{-\infty}^{t} e^{-1000\tau} d\tau$$

- Sabemos que $h(t)$ inicia a partir do 0:

$$y_{coringa}(t) = \int_{0}^{t} e^{-1000\tau} d\tau$$

$$y_{coringa}(t) = \left[ \frac{e^{-1000\tau}}{-1000} \right]_{0}^{t}$$

* Aplicando os intervalos:

$$y_{coringa}(t) = -\frac{1}{1000} (e^{-1000t} - e^{0})$$

$$\boxed{y_{coringa}(t) = \frac{1 - e^{-1000t}}{1000}}$$

- Aplicando os deslocamentos em $y_{coringa}(t)$:

* ***Termo associado a $u(t + 0,0025)$:***
$$\boxed{y_{coringa}(t + 0,0025) = \left( \frac{1 - e^{-1000(t + 0,0025)}}{1000} \right) u(t + 0,0025)}$$

* ***Termo associado a $u(t - 0,0025)$:***
$$\boxed{y_{coringa}(t - 0,0025) = \left( \frac{1 - e^{-1000(t - 0,0025)}}{1000} \right) u(t - 0,0025)}$$

- Substituindo na expressão de $y(t)$:

$$y(t) = 4000 \left[ \left( \frac{1 - e^{-1000(t + 0,0025)}}{1000} \right) u(t + 0,0025) - \left( \frac{1 - e^{-1000(t - 0,0025)}}{1000} \right) u(t - 0,0025) \right]$$

* Colocando o 1000 para fora:

$$y(t) = \frac{4000}{1000} \left[ (1 - e^{-1000(t + 0,0025)}) u(t + 0,0025) - (1 - e^{-1000(t - 0,0025)}) u(t - 0,0025) \right]$$

$$y(t) = 4 \left[ (1 - e^{-1000(t + 0,0025)}) u(t + 0,0025) - (1 - e^{-1000(t - 0,0025)}) u(t - 0,0025) \right]$$

$$\boxed{y(t) = 4(1 - e^{-1000(t + 0,0025)}) u(t + 0,0025) - 4(1 - e^{-1000(t - 0,0025)}) u(t - 0,0025)}$$

- $x(t)$ é um pulso retangular de 5 ms:

Olhando matematicamente, a saída $y(t)$ é mais suave à medida que a cauda é maior, e mais abrupta (parecida com o pulso) se a cauda é menor. Olhando fisicamente, é como se $h(t)$ determinasse quanto tempo o sistema leva para carregar e descarregar:

# ***Analisando no domínio da frequência:***

- $X(\omega)$ é a função sinc. Como a entrada é um pulso retangular, há muita energia em volta de $\omega=0$ (frequências mais baixas) e pouca energia em frequências maiores que se estendem para o infinito (formam as quinas do retângulo).

- Podemos ver que $H(\omega)$ é um filtro passa-baixa com corte em 1000 rad/s. Podemos confirmar isso da seguinte forma:

  $$H(\omega) = \frac{1000}{j\omega + 1000}$$

  - Para $\omega=0$, ficamos com:

  $$H(0) = \frac{1000}{1000} = 1$$

  - Ou seja, para frequências baixas o ganho do sinal é 1. Para $\omega=1000$:

  $$H(1000) = \frac{1000}{j1000 + 1000}$$

  - Calculando o módulo:

  $$|H(1000)| = \frac{1000}{\sqrt{1000^2 + 1000^2}}$$

  $$|H(1000)| \approx 0,707$$

  - Dessa forma, para 1000 rad/s, somente 70,7% do sinal original passa, onde essa porcentagem cai ainda mais à medida que aumentamos a frequência a partir de 1000 rad/s.

- A saída $Y(\omega)$, como esperado, se parece com a função sinc, visto que a área do sinal também está concentrada em frequências próximas de $\omega=0$, mas com frequências altas atenuadas, já que a saída foi arredondada em $h(t)$ e filtrada em $H(\omega)$.
